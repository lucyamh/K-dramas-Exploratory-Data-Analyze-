> 2. HOW MANY DRAMAS HAVE BEEN RELEASED PER MONTH?



```python
#First, I will split the Aired Date column into Start_date and End_date
data["Aired Date"].str.split("-", expand=True)
```

```python
#Then put it in into two new columns. Those name are Start_date and End_date
data[["Start_date","End_date"]] = data["Aired Date"].str.split("-", expand=True)
```

```python
#After that, I remove the leading white space of the start_date column
data["Start_date"] = data["Start_date"].str.strip()
```

```python
data["Start_date"].dtype
```

```python
#As we know, the start_date column has Object format so I should change it into datetime format
data["Date"] = pd.to_datetime(data["Start_date"], format="%b %d, %Y")
```

```python
#Because I want to know the number of drama per MONTH so I will create Month column from the date column
data["Month"] = data["Date"].dt.strftime("%B")
```

```python
data.head(2)
```

```python
ax = sns.countplot(x="Month", data=data, color='Blue')
ax.bar_label(ax.containers[0], fmt= '%0.1f', label_type='edge')
plt.xticks(rotation=45)
plt.title("THE NUMBER OF DRAMAS PER MONTH")
```

6. FIND THE MOST OF DURATION IN THE DRAMAS

```python
data["Duration"].dtype
```

```python
data["Duration"].value_counts()
```

```python
#change Duration column into integer format
```

```python
duration = []
for i in range(len(data["Duration"].values)):
  if "1 hr." in data["Duration"].values[i]:
    duration.append(int(data["Duration"].values[i].split()[2])+60)
  else:
    duration.append(int(data["Duration"].values[i].split()[0]))
data["Duration"] = duration
data.head(3)
```

```python
data["Duration"].value_counts().sort_index()
```

```python
def duration(duration):
  if duration >=60:
    return 0
  else:
    return 1
```

```python
data["time_duration"] = data["Duration"].apply(duration)
```

```python
data.head()
```

```python
ax=sns.countplot(x='time_duration',data=data, color='Blue')
ax.bar_label(ax.containers[0], fmt= '%0.1f', label_type='edge')
plt.title("THE NUMBER OF DRAMAS IN TIME_DURATION CATEGORY")
```

```python
#from the graph we know that time_duration of the dramas. Number 0 has the most of dramas than the number 1.
#It means the most of dramas has the duration over 60 minutes than under 60 minutes
```

> 7. HOW MANY EPISODES IN DRAMAS?



```python
def episode(episode):
  if episode >=20:
    return 0
  elif episode >=10:
    return 1
  else:
    return 2
```

```python
data["episode_code"] = data["Number of Episodes"].apply(episode)
```

```python
data.head(3)
```

```python
ax=sns.countplot(x="episode_code",data=data, color='Blue')
ax.bar_label(ax.containers[0], fmt= '%0.1f', label_type='edge')
plt.title("THE NUMBER OF DRAMAS IN EPISODE CATEGORY")
```

```python
#as we see 10-20 episodes has the most value
#it means the dramas have 10-20 episodes in average.
```
